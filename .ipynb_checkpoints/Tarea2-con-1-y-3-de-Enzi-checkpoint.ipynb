## 1. Tipos de fronteras en clasificación

Como se ha discutido en clases, los problemas de clasificación pueden ser representados de distintas maneras, en donde estas representaciones definen un espacio de entrada del dominio de los datos ($\mathbb{R}_{\mathbb{X}}$). Los ejemplos dentro del espacio de entrada (*manifold*) pueden tener distintas formas, donde estas indicarán si es que estos ejemplos tendrán fronteras lineales o no. Con el propósito de analizar los distintos tipos de fronteras que definen los algoritmos de clasificación, se trabajará con un dataset sintético lo cual es ideal para analizar los diferentes tipos, así hacer un recorrido rápido por las grandes ideas en la búsqueda dela no-linealidad. Este dataset de jugete está en un espacio de 2 dimensiones y es conformado por dos conjuntos de datos, pudiendo ver que la frontera entre ellos claramente no es lineal. El dataset representa al problema XOR, pero para trabajarlo de una manera distinta se generán gausianas para agregar un poco de ruido en el conjunto de entrenamiento y que no sea un problema trivial. El código que los genera es el siguiente:

````python
import numpy as np
def do_gaussianXOR(n1=250,n2=250,std= 0.3):
    rng = np.random.RandomState(0) #crea un generador pseudo aleatorio con semilla igual a cero
    u = 0.5
    #el ruido se agrega aca
    #generar distribuciones normales en cada cuadrante con ruido haciendo que algunos puntos se escapen
    primer_c = rng.normal(loc=[u,u],scale=std,size=(n1,2))
    segundo_c = rng.normal(loc=[-u,u],scale=std,size=(n2,2))
    tercer_c = rng.normal(loc=[-u,-u],scale=std,size=(n1,2))
    cuarto_c = rng.normal(loc=[u,-u],scale=std,size=(n2,2))
    #X_train se crea una matriz de N,2 donde se tienen N puntos con coordenadas X e Y 
    X_train = np.vstack((primer_c,segundo_c,tercer_c,cuarto_c))
    #Y_train es un vector con las etiquetas de los N puntos donde primer y tercer cuadrante son 0 y 2do y 4to son 1
    Y_train = np.hstack((np.zeros(n1),np.ones(n2),np.zeros(n1),np.ones(n2)))
    #X_test es una uniforme entre -1 y 1 de 500 datos y 2 columnas
    X_test = rng.uniform(low=-1.0,high=1.0,size=( (n1+n2),2))
    #Y_test es el XOR de los originales de X_test 
    Y_test = np.logical_xor(X_test[:,0] >= 0, X_test[:,1] >= 0) #consideramos 0 como positivo
    return X_train,Y_train,X_test,Y_test,primer_c,segundo_c,tercer_c,cuarto_c
````

Para contextualizar, el $XOR(a,b)$ es una operación lógica que retorna 1 si $a$ y $b$ son distintos y 0 si no. Dado en el contexto actual, un plano cartesiano, se va a tener que el primer y tercer cuadrante deben ser etiquetados con 0 ya que ambas componentes tienen el mismo signo y el segundo y cuarto cuadrante serán marcados con 1 ya que tienen signos distintos.

> b) Entrene el clasificador Linear Discriminant Analysis (LDA) y visualice la frontera de decisión que define este algoritmo. Comente cualitativamente lo que observa.

````python
#Código para visualizar las fronteras de un modelo
import matplotlib.pyplot as plt
def visualize_border(model,x,y,title=""):
    fig, axis = plt.subplots(1, 1,figsize=(12, 6))

    plt.scatter(x[:,0], x[:,1], s=50, c=y, cmap=plt.cm.winter)
    h = .02 # step size in the mesh (es como la smoothness de la linea)
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1 #generan limites del rectangulo/cuadrado
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1 #x[0] eje x, x[1] eje y 
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]) #or predict_proba
    Z = Z.reshape(xx.shape)
    Zplot = Z > 0.5
    
    axis.scatter(x[:,0],x[:,1],s=50,c=y,zorder=10,cmap=plt.cm.winter)
    axis.contour(xx, yy, Z,levels=[-1, 0, 1])
    axis.pcolormesh(xx, yy, Zplot ,cmap=plt.cm.YlGn)
    plt.title(title)
    plt.show()
````

````python
import warnings
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
warnings.filterwarnings("ignore")
model = LDA(solver='lsqr')
model.fit(X_train2,Y_train2)
visualize_border(model,X_train2,Y_train2,"LDA con error 0.1 en el conjunto train")
print("LDA score en conjunto test: {}".format(model.score(X_test2,Y_test2)))
print("Matriz de covarianza LDA:\n {}".format(model.covariance_))
model.fit(X_train3,Y_train3)
visualize_border(model,X_train3,Y_train3,"LDA con error 0.5 en el conjunto train")
print("LDA score en conjunto test: {}".format(model.score(X_test3,Y_test3)))
print("Matriz de covarianza LDA:\n {}".format(model.covariance_))
model.fit(X_train,Y_train)
visualize_border(model,X_train,Y_train,"LDA con error 0.3 en el conjunto train")
visualize_border(model,X_test,Y_test,"LDA con error 0.3 en el conjunto test")
print("LDA score en conjunto test: {}".format(model.score(X_test,Y_test)))
print("Matriz de covarianza LDA:\n {}".format(model.covariance_))
````

Del entrenamiento del clasificador LDA se puede observar que la frontera de decisión que genera es la misma para cualquier conjunto de datos utilizado, lo que llama la atención es que hay una separación en dos cuadrantes y no en cuatro y en cada separación tenemos tantos datos calificados con 0 como calificados con 1. 

Cabe destacar que a medida que mientras más separados los datos menor son los coeficientes de la diagonal de LDA y a medida que se juntan más, como en el caso del ruido 0.5 se tiene que la matriz de covarianza presenta valores más altos en la diagonal.

>c) Entrene el clasificador Quadratic Discriminant Analysis (QDA) y visualice la frontera de decisión que define este algoritmo. Analice cualitativamente lo que observa y compare con LDA, en qué difieren y en qué se asemejan ¿Qué  distribución de probabilidad asume cada uno?

````python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
warnings.filterwarnings("ignore")
model = QDA(store_covariance = True)
model.fit(X_train2,Y_train2)
visualize_border(model,X_train2,Y_train2,"QDA con error 0.1 en el conjunto train")
model.fit(X_train3,Y_train3)
visualize_border(model,X_train3,Y_train3,"QDA con error 0.5 en el conjunto train")
model.fit(X_train,Y_train)
visualize_border(model,X_train,Y_train,"QDA con error 0.3 en el conjunto train")
visualize_border(model,X_test,Y_test,"QDA con error 0.3 en el conjunto test")
print("QDA score en conjunto test: {}".format(model.score(X_test,Y_test)))
print("Matrices covarianza de QDA: \n{0} \n y \n {1}".format(model.covariance_[0],model.covariance_[1]))
````

De estos gráficos se puede observar que QDA, a diferencia de LDA, separa el conjunto en 4 partes donde cada parte representa casi un cuadrante del plano y se puede ver que para el conjunto test el score es más del doble que en LDA. La única similitud entre ambos modelos es que los dos tienen una división vertical, aunque en QDA no es una linea exactamente si no que 2 parábolas cuyos vérticas están bastante cercanos dando la ilusión de un línea vertical.

Ambos modelos se generan al ajustar los datos utilizando la regla de bayes:

$$ 
P(y=k|X) = \frac{P(X|y=k)\cdot P(y=k)}{P(X)} = \frac{P(X|y=k)\cdot P(y=k)}{\sum_{l}P(X|y=l)\cdot P(y=l)}
$$

Ambos métodos consideran un modelo gaussiano multivariado para $P(X|y)$:

$$
P(X|y=k) = \frac{1}{\left(2\pi\right)^{d/2}\cdot|\Sigma_{k}|^{2}}\cdot exp\left(-\frac{1}{2}\left(X - \mu_{k} \right)^{T}\Sigma_{k}^{-1}\left(X - \mu_{k} \right) \right)
$$

Donde $\Sigma$ es la matriz de covarianza, la diferencia entre estos dos métodos es que LDA asume esta matriz como constante para todas las clases lo que lleva a fronteras de decisión lineales mientras que QDA no realiza ese supuesto y esto deriva en las fronteras de decisión cuadráticas. 

A partir de los scores de cada técnica se puede decir que QDA tiene un desempeño mucho mejor que LDA podemos atribuir esto a que las matrices de covarianza no son las mismas como se puede ver, la precisión es un poco más de el doble en este caso con el set de validación.

Una SVM Lineal lo que hace es determinar un margen lineal (como el nombre indica) que separe el conjunto de datos. La idea es encontrar el hiperplano separador que maximice el margen entre las dos clases como la siguiente imagen ilustra:

<img src="https://qph.fs.quoracdn.net/main-qimg-675fedee717331e478ecfcc40e2e4d38" title="Hard Margin SVM" width="65%" height="65%" />

Si la SVM es de un margen rígido esto quiere decir que no se tolerarán datos mal clasificados y los *support vectors* serán los puntos en el hiperplano más cercano al hiperplano separador, en el caso de la imagen serían los puntos remarcados sobre las líneas. El grave problema de esta SVM es que si el problema no es linealmente separable no encuentra una solución al problema.

De lo anterior se desglosa una SVM de margen flexible que permite cierta holgura en el margen, permitiendo que este se expanda tolerando puntos en la zona entre la frontera de decisión y el margen definido, además de datos mal etiquetados.

<img src="https://i.stack.imgur.com/npEOk.png" title="Soft Margin SVM" width="65%" height="65%"/>

Sin embargo, como se mencionó con anterioridad, se tiene que los puntos de una misma clase se ubican en un cuadrante y en el que se encuentra opuesto por el vértice por así decirlo. Esta situación genera que el problema no sea linealmente separable por lo que una SVM lineal no debería de poder resolver el problema, esto se verá a continuación donde se utilizarán varios valores para la holgura del márgen en una SVM lineal.

El problema de optimización relacionado a la SVM, en su versión de margen flexible, corresponde al siguiente:

$$
min \frac{1}{2}\mid\mid w \mid\mid ^{2} + C\cdot \sum_{i=1}^{n}\xi_{i} 
$$

Sujeto a (n restricciones)

$$
y_{i}\left(w^{T}x+b \right)\geq 1+\xi_{i}
$$

Donde $\xi$ son las holguras asociadas a los puntos para cumplir las restricciones y $C$ un parámetro de regularización de las holguras.

````python
from sklearn.svm import SVC as SVM #SVC is for classification
from ipywidgets import interactive
def train_model(param):
    model= SVM()
    C = 10**(param) #penalizador de los slacks de SVM 
    print("Param C= ",C)
    model.set_params(C=C,kernel='linear')
    model.fit(X_train,Y_train)
    print("Test Accuracy in training set: ",model.score(X_train,Y_train))
    print("Test Accuracy in validation set: ",model.score(X_test,Y_test))
    return model
def visualize_border_interactive(param):
    model = train_model(param)
    #visualize_border(model,X_train,Y_train,"Entrenamiento en conjunto entrenamiento con parametro " + str(param))
    visualize_border(model,X_test,Y_test,"Entrenamiento en conjunto validación con parametro " + str(param))
p_min = -10 #
p_max = 4 #No subir de 4, se queda pegado calculando 
interactive(visualize_border_interactive,param=(p_min,p_max))
````

Como es de esperar, una SVM lineal no es capaz de separar este problema debido a la distribución espacial de las clases asociadas a este a pesar de que se varíe el parámetro de regularización.

A diferencia de la SVM linear, una SVM no linear permite delimitar fronteras no lineales lo que permite obtener márgenes más flexibles y que podrían permitir resolver el problema. Esto se puede lograr mediante una extensión vía kernels, en este sección se verán 2 kernels: polinomial y rbf cuyas expresiones son las siguientes: 

**RBF o kernel gaussiano**

$$
k\langle x,x'\rangle = exp(-\gamma \mid\mid x-x' \mid\mid ^{2})
$$

Donde $\gamma$ (asumido positivo) ajusta la anchura de la gaussiana, notar que valores altos de $\gamma$ harán tender la expresión a $0$ o si $\mid\mid x-x' \mid\mid ^{2} \gg \frac{1}{\gamma}$

**Polinomial**

$$
k\langle x,x'\rangle = \left(\alpha\cdot x^{T}x + r\right)^{d}
$$

Donde $\alpha$ es una constante multiplicativa que para efectos de este trabajo será $1$, $r$ una constante aditiva que será $0$ y $d$ el grado del polinomio

````python
from sklearn.svm import SVC as SVM #SVC is for classification
from ipywidgets import interactive
def train_model_rbf(param,gam):
    model= SVM()
    C = 10**(param) #penalizador de los slacks de SVM 
    print("Param C= ",C)
    model.set_params(C=C,kernel='rbf',gamma=1/gam) #parámetro inverso 
    model.fit(X_train,Y_train)
    print("Test Accuracy in training set: ",model.score(X_train,Y_train))
    print("Test Accuracy in validation set: ",model.score(X_test,Y_test))
    return model
def visualize_border_interactive_rbf(param,gamma):
    model = train_model_rbf(param,gamma)
    #visualize_border(model,X_train,Y_train,"SVM con kernel RBF con gamma {0} y parametro regularizador {1} en train set".format(1/gamma,10**param))
    visualize_border(model,X_test,Y_test,"SVM con kernel RBF {0} y parametro regularizador {1} en validation set".format(1/gamma,10**param))
p_min = -10 #
p_max = 4 #No subir de 4, se queda pegado calculando 
g_min = 1
g_max = 10
interactive(visualize_border_interactive_rbf,param=(p_min,p_max),gamma=(g_min,g_max))
````

````python
def train_model_poly(param,grado):
    model= SVM()
    C = 10**(param) #penalizador de los slacks de SVM 
    print("Param C= ",C)
    model.set_params(C=C,kernel='poly',degree=grado,gamma=1,coef0=0)
    model.fit(X_train,Y_train)
    print("Test Accuracy in training set: ",model.score(X_train,Y_train))
    print("Test Accuracy in validation set: ",model.score(X_test,Y_test))
    return model
def visualize_border_interactive_poly(param,grado):
    model = train_model_poly(param,grado)
    #visualize_border(model,X_train,Y_train,"SVM con kernel Polinomial grado {0} y parametro regularizador {1} en train set".format(grado,10**param))
    visualize_border(model,X_test,Y_test,"SVM con kernel Polinomial grado {0} y parametro regularizador {1} en validation set".format(grado,10**param))
p_min = -10 #
p_max = 3 #No subir de 3, se queda pegado calculando 
grado_min = 1
grado_max = 8
interactive(visualize_border_interactive_poly,param=(p_min,p_max),grado=(grado_min,grado_max))
````

Al utilizar kernels no lineales (polinomial con grado mayor a uno y rbf) se tienen soluciones que moldean la frontera acorde a la distribución de los datos y tienden a agrupar los dos conjuntos en distintas partes de la frontera. Se pueden evidenciar mejores resultados utilizando un kernel rbf que un kernel polinomial. También se destaca que cuando el kernel polinomial es de grado 1 se obtiene la frontera lineal de LDA y cuando es de grado 2 se obtiene la frontera lineal de QDA, lo que tiene sentido dados los grados de estos dos métodos. También se observa que las fronteras varían ligeramente si se utiliza el conjunto de entrenamiento y el de validación.

Un tipo de problema un poco diferente a la clásica clasificación que hemos visto es el de etiquetación múltiple o *multi-label* [[5]](#refs2), en donde se debe asignar una *o un grupo* de etiquetas a un dato, es decir, puede ser tratado como un problema de clasificación donde las clases pasan a ser **NO** mutuamente excluyente, es decir puede ser asignada más de una a un dato. Este problema se ve muchas veces en imágenes que contienen más de algun objeto, por lo que las clases pasan a ser no excluyente, o en sí se transforma en un problema de *multi-label*. También aparece en texto cuando se debe asignar un tópico a un documento, ya que podría hablar de diferentes cosas, como religión, política, finanzas o educación, por lo que podría contener varios.

````python
def evaluation_metric(y_true,y_pred):
    suma = 0.
    for y,yp in zip(y_true,y_pred):
        suma += np.dot(y,yp) / np.sum(y)  #cantidad de valores iguales
    return suma/float(len(y_true))
````

La métrica presentada solo evalua la performance a partir de los positivos verdaderos y los falsos negativos, ignorando el otro tipo de error del clasificador: los falsos positivos. Estos errores pueden ser o no importantes dependiendo del uso que se le de a los resultados obtenidos, como el caso de los autos sin conductor, donde un falso positivo o negativo puede resultar en un accidente. Además, se ve altamente afectado si la cantidad de etiquetas verdaderas es 0, generando una división por cero.
Si se desea tener una métrica que considere también los falsos positivos, se propone la siguiente métrica:

````python
def evaluation_metric2(y_true,y_pred):
    suma = 0.
    for y,yp in zip(y_true,y_pred):
        suma += np.sum(y == yp, dtype='float') / len(y)  #cantidad de valores iguales
    return suma/float(len(y_true))
````

## 3. *Taxi Trip Duration*: Predicción de duración de viaje

<img src="https://upload.wikimedia.org/wikipedia/commons/4/4d/Newyorkcity_taxi_motion_banner.jpg" />


En esta actividad trabajaremos con un dataset de una competencia en la plataforma Kaggle, el cual se llama **New York City Taxi Trip Duration** y que puede ser accedido a través del siguiente __[link](https://www.kaggle.com/c/nyc-taxi-trip-duration/)__. Como se puede ver es una competencia para predecir la duración del viaje de un taxi en Nueva York, por lo que es un problema de regresión que resulta altamente no lineal  y no determinista debido a las invariantes que tiene, por ejemplo un viaje de un punto a otro puede durar distinto tiempo dependiendo de múltiples factores, como el horario, la congestión de ese día u otros.

La métrica de evaluación de la competencia y la que utlizaremos será el RMSLE (*Root Mean Squared Logarithm Error*):

$$
E(y,\hat{y}) = \sqrt{ \frac{1}{n} \sum_{i=1}^n \left( \log{(y_i +1)}- \log{(\hat{y}_i+1}) \right)^2}
$$

Sobre RMSLE:

Este error no es tan popular como MAE o MSE, pero la utilidad de este error es cuando la variable objetivo (predicción) posee un espectro amplio de valores y así no se penalizan los errores grandes tan severamente al utilizar las variables transformadas por un logaritmo.

````python
import numpy as np
#Error
def rmsle(predicted,real):    
    suma=0.0
    for x in range(len(predicted)):
        p = np.log(predicted[x]+1)
        r = np.log(real[x]+1)
        suma += np.power((p - r),2)
    return np.sqrt(suma/len(predicted))
````

Ahora que se ha hecho el filtrado de datos se preprocesará lo que queda. El preprocesamiento consiste en extraer características que puedan resultar interesantes y tengan algún peso en la duración del viaje como las listadas a continuación: 

+ Determinar la hora de inicio del viaje
+ Determinar el día 0:lunes....6:domingo
+ Determinar el mes 1:enero....12:diciembre
+ Si el viaje fue en un feriado o no (holiday)
+ La estación del año en que se hizo el viaje 

Ya que estos valores son útiles a la hora de predecir el largo de un viaje, por ejemplo, en invierno se podrían realizar viajes más largos de gente que quiera evitar la lluvia un día de semana en la tarde, u horario de salida del trabajo.

Para ésto la librería *keras* nos ayuda:

````python
from keras.utils import to_categorical
import numpy as np
y_onehot = to_categorical(y_train) #argumento num_classes es automatico, fuerza que el vector sea de rango 0 a num_clases
````

Como se menciono anteriormente, la funcion loss *Categorical C-E* es útil para problemas de categorización multi-clase, lo cual aplica el modelo *SoftMax* en la capa de salida. Se probará el desempeño de la red con K-cross validation con K=10.

````python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
model = Sequential()
model.add(Dense(256, input_dim=X_train.shape[1], activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(y_onehot.shape[1], activation="sigmoid"))
model.compile(optimizer=SGD(lr=0.1), loss="categorical_crossentropy")
````

````python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import KFold, cross_val_score

np.random.seed(70)
estimator = KerasClassifier(build_fn=model, epochs=40, batch_size=128, verbose=1)
kfold = KFold(n_splits=10, shuffle=True, random_state=70)

results = cross_val_score(estimator, X_train, y_onehot, cv=kfold)
print (results)


````
