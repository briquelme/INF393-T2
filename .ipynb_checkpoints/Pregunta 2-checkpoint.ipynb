## 2. Un problema* Multi-label*

<img src="https://files.virgool.io/upload/users/12226/posts/oxo6jvmweqmr/1ios2rcvohwc.png" width="35%" style="float: right;" />

Un tipo de problema un poco diferente a la clásica clasificación que hemos visto es el de etiquetación múltiple o *multi-label* [[5]](#refs2), en donde se debe asignar una *o un grupo* de etiquetas a un dato, es decir, puede ser tratado como un problema de clasificación donde las clases pasan a ser **NO** mutuamente excluyente, es decir puede ser asignada más de una a un dato. Este problema se ve muchas veces en imágenes que contienen más de algun objeto, por lo que las clases pasan a ser no excluyente, o en sí se transforma en un problema de *multi-label*. También aparece en texto cuando se debe asignar un tópico a un documento, ya que podría hablar de diferentes cosas, como religión, política, finanzas o educación, por lo que podría contener varios.


En esta actividad se trabajará con una versión liviana del dataset __[**NUS-WIDE**](https://lms.comp.nus.edu.sg/research/NUS-WIDE.htm)__ , el cual consta de imágenes extraídas de la web (Flickr.com) en donde los diferentes *tags* o conceptos asociados a la imagen pasan a ser las etiquetas/*labels*. De la página donde se alberga NUS-WIDE será necesario descargar:
* Concept List: http://dl.nextcenter.org/public/nuswide/ConceptsList.zip
* NUS-WIDE-Lite: http://dl.nextcenter.org/public/nuswide/NUS-WIDE-Lite.rar
* NUS-WIDE-Urls: http://dl.nextcenter.org/public/nuswide/NUS-WIDE-urls.rar

````python
import pandas as pd
import numpy as np
y = pd.read_csv("NUS-WIDE-Lite/NUS-WIDE-Lite_groundtruth/Lite_GT_Train.txt",
                delimiter="\s",
                header=None, 
                engine='python').values
print(y.shape)
print("Ejemplos por clase:")
print(y.sum(axis=0))
````

El archivo contiene 27807 lineas, de las cuales cada una contiene 81 números binarios de 1 dígito. Se presume que este archivo describe cuales de las 81 etiquetas fueron asignadas a cada uno de ellos, siendo representados por cada linea del archivo.

Para verificar esta suposición, se procede a abrir los archivos `Lite_Labels_Train` dentro del directorio, y formar con ellos un array similar al archivo inicialmente abierto.

````python
import glob, os
folder_dir = "NUS-WIDE-Lite/NUS-WIDE-Lite_groundtruth/"
file_names = glob.glob(folder_dir+"Lite_Labels_*_Train.txt")
file_names.sort()
labels = [pd.read_csv(fn,
                delimiter="\s",
                header=None, 
                engine='python').values
         for fn in file_names]
````

````python
labels_array = np.concatenate(labels, axis=1)
print(np.shape(y == labels_array))
````

Como ambos array generados son identicos, se puede asumir que los nombres de las etiquetas de las categorías presentes del archivo `Lite_GT_Train.txt` corresponden a los nombres de las etiquetas presentados en los otros archivos en orden alfabético. Con esta información se puede componer el dataframe de las etiquetas de entrenamiento.

````python
import re
labels_head = [
    re.sub(r'NUS-WIDE-Lite\/NUS-WIDE-Lite_groundtruth\/Lite_Labels_(\w*)_Train.txt',
           r'\1',
           fn)
    for fn in file_names
]
labels_train = pd.DataFrame(y,
                         columns = labels_head)
print(labels_train.shape)
labels_train.head()
````

````python
import re
from urllib.request import urlopen
label_names = np.loadtxt("ConceptsList/Concepts81.txt",dtype='object')
arch = open("NUS-WIDE-urls.txt",'r') #lectura para mapeo de link a url
regex = re.compile("\s+")
image2url = {}
for line in arch.readlines():
    aux = regex.split(line.strip())
    image = aux[0].split("Flickr\\")[-1]
    url_m = aux[3]
    image2url[image] = url_m
arch.close()
arch = open("NUS-WIDE-Lite/image list/Train_imageOutPutFileList.txt",'r')
data_urls = [image2url[line.strip()] for line in arch.readlines()]
arch.close()
````

Que una imágen posea múltiples etiquetas solo indica la presencia de múltiples elementos capturados en dicha imágen, lo cual puede ser desde 0 hasta la cantidad de etiquetas conocidas para una imágen en específico. Esto se puede apreciar con el output del código anterior.  
En contraste a la clasificación multiclase, la cual indica si el objeto pertenece a solo una de las N clases conocidas (posiblemente, teniendo como output un entero entre 1 y N), la clasificación multi-label busca asignar a un objeto todas las clases al que corresponde, dentro de un universo de N clases (entregando un vector de N binarios). Usualmente este método se usa para asignar a un objeto varias clases que no son excluyentes entre sí (por ejemplo, géneros de un libro), mientras que la clasificación multiclase se enfoca en clasificaciones excluyentes (por ejemplo, a que tipo corresponde una fruta).

````python
def evaluation_metric(y_true,y_pred):
    suma = 0.
    for y,yp in zip(y_true,y_pred):
        suma += np.dot(y,yp) / np.sum(y)  #cantidad de valores iguales
    return suma/float(len(y_true))
````

La métrica presentada solo evalua la performance a partir de los positivos verdaderos y los falsos negativos, ignorando el otro tipo de error del clasificador: los falsos positivos. Estos errores pueden ser o no importantes dependiendo del uso que se le de a los resultados obtenidos, como el caso de los autos sin conductor, donde un falso positivo o negativo puede resultar en un accidente. Además, se ve altamente afectado si la cantidad de etiquetas verdaderas es 0, generando una división por cero.
Si se desea tener una métrica que considere también los falsos positivos, se propone la siguiente métrica:

````python
def evaluation_metric2(y_true,y_pred):
    suma = 0.
    for y,yp in zip(y_true,y_pred):
        suma += np.sum(y == yp) / len(y)  #cantidad de valores iguales
    return suma/float(len(y_true))
````

De manera nativa, ``DecisionTreeClassifier`` soporta de manera nativa problemas con salida múltiple[[5]](#refs2), lo cual corresponde a la clasificación multi-label. Como las labels de los conjuntos de prueba y entrenamiento ya estan binarizados, no es necesario realizarles modificaciones previas.  

````python
%%time
from sklearn.tree import DecisionTreeClassifier as Tree
#criterios gini(sin logs, reduce misclass) y entropy (analisis exploratorio)
#splitter best o random
#max_Depth corresponde a niveles, juega con esto
models_best_gini = [Tree().set_params(max_depth=depth,criterion="gini",splitter="best").fit(X_train,y_train) for depth in [1, 5, 10, 20, 30, 50]]
````

````python
%%time
models_best_entr = [Tree().set_params(max_depth=depth,criterion="entropy",splitter="best").fit(X_train,y_train) for depth in [1, 5, 10, 20, 30, 50]]
````

````python
%%time
models_rand_gini = [Tree().set_params(max_depth=depth,criterion="gini",splitter="random").fit(X_train,y_train) for depth in [1, 5, 10, 20, 30, 50]]
````

````python
%%time
models_rand_entr = [Tree().set_params(max_depth=depth,criterion="entropy",splitter="best").fit(X_train,y_train) for depth in [1, 5, 10, 20, 30, 50]]
````

Como se puede apreciar con la cell magic ``%%time``, el criterio de entropía toma mayor tiempo de ejecución, debido a que emplea cálculos de logaritmos, lo cual lo hace mas costoso computacionalmente. Debido a esto y que los resultados son similares a los obtenidos con el índice de Gini tienden a ser idénticos, se recomienda emplear este último criterio.

````python
#predicciones train
y_pred_train_b_g = [model.predict(X_train) for model in models_best_gini]
y_pred_train_b_e = [model.predict(X_train) for model in models_best_entr]
y_pred_train_r_g = [model.predict(X_train) for model in models_rand_gini]
y_pred_train_r_e = [model.predict(X_train) for model in models_rand_entr]
#predicciones test
y_pred_test_b_g = [model.predict(X_test) for model in models_best_gini]
y_pred_test_b_e = [model.predict(X_test) for model in models_best_entr]
y_pred_test_r_g = [model.predict(X_test) for model in models_rand_gini]
y_pred_test_r_e = [model.predict(X_test) for model in models_rand_entr]
````

````python
#errores entrenamiento
#metrica 1
eval_train_b_g = [evaluation_metric(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_b_g]
eval_train_b_e = [evaluation_metric(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_b_e] 
eval_train_r_g = [evaluation_metric(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_r_g] 
eval_train_r_e = [evaluation_metric(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_r_e] 
#metrica 2
eval2_train_b_g = [evaluation_metric2(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_b_g]
eval2_train_b_e = [evaluation_metric2(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_b_e] 
eval2_train_r_g = [evaluation_metric2(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_r_g] 
eval2_train_r_e = [evaluation_metric2(y_true=y_train, y_pred=prediction) for prediction in y_pred_train_r_e] 
````

````python
print (eval_train_b_g)#metrica entregada
print (eval2_train_b_g)#metrica propuesta
````

Como se puede apreciar, la métrica entregada falla debido a la existencia de *divide by zero* dentro del cálculo. Podría repararse dicha métrica al forzar que entrege el valor 1 a la suma, ya que en teoría no tendria falsos negativos.  
La métrica sugerida presenta bajo error de entrenamiento usando Gini y particiones óptimas locales. Desde aquí en adelante se usará la última métrica para evitar el error matemático de la primera.  

````python
import matplotlib.pyplot as plt

plt.plot(eval2_train_b_g, label="Best-Gini.")
plt.plot(eval2_train_b_e, label="Best-Entropy")
plt.plot(eval2_train_r_g, label="Random-Gini")
plt.plot(eval2_train_r_e, label="Random-Entropy")
plt.title("Accuracy v/s Max. Depth")
plt.ylabel("Accuracy")
plt.xlabel("Max Depth")
plt.xticks(range(0,6), [1, 5, 10, 20, 30, 50])
plt.legend()
plt.show()
````

````python
eval2_test_b_g = [evaluation_metric2(y_true=y_test, y_pred=prediction) for prediction in y_pred_test_b_g]
eval2_test_b_e = [evaluation_metric2(y_true=y_test, y_pred=prediction) for prediction in y_pred_test_b_e] 
eval2_test_r_g = [evaluation_metric2(y_true=y_test, y_pred=prediction) for prediction in y_pred_test_r_g] 
eval2_test_r_e = [evaluation_metric2(y_true=y_test, y_pred=prediction) for prediction in y_pred_test_r_e] 
````
